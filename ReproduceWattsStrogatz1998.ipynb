```python
import numpy as np 
import networkx as nx 
import matplotlib.pyplot as plt 
from mpl_toolkits.axes_grid1 import make_axes_locatable
import scipy.io
from scipy.sparse import csr_matrix
%matplotlib inline

#############
#plot figure1
#############
node = 20
k = 4
P = [0,0.5,1]
plt.figure(figsize=(20,6))
for i in range(3):
    tmp = 131+i
    rgl_latt=nx.watts_strogatz_graph(node,k,P[i])
    pos = nx.circular_layout(rgl_latt)
    plt.subplot(str(tmp))
    plt.title('p = '+(str(P[i])))
    nx.draw_networkx(rgl_latt,node_size=50,style='dashed',width=0.3,pos=pos)
plt.savefig("watts_strogatz_figure_1.eps")

```

```python

#############
#plot figure2
#############
node = 1000
k = 10
L = []
C = []
sample = 30
P = np.logspace(-4,0,sample)
for i in range(sample):
    rgl_latt = nx.watts_strogatz_graph(node,k,P[i])
    L.append(nx.average_shortest_path_length(rgl_latt))
    C.append(nx.average_clustering(rgl_latt))
plt.figure(figsize=(6,6))
plt.semilogx(P,np.asarray(L)/L[0],'bo--',label = 'L')
plt.semilogx(P,np.asarray(C)/C[0],'rv--',label = 'C')
plt.legend()
plt.savefig("watts_strogatz_figure_2-2.eps")

plt.show()
```
